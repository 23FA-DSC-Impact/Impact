```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Read Data

```python
dataset = pd.read_csv(os.path.join(os.getcwd(), "dataset", "prepared", "dataset.csv"),
                      sep=",",
                      header=0,
                      dtype={"date": object, "close": float, "open": float,
                             "high": float, "low": float,
                             "gdp": float, "effr": float,
                             "acwi_adj_close": float})
dataset['date'] = pd.to_datetime(dataset['date'])
dataset.info()
```

# Visualize Data

```python
dataset.columns
```

```python
# Plot 1: Close, Open, High, Low
df = dataset

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label='Close')
plt.plot(df['date'], df['open'], label='Open')
plt.plot(df['date'], df['high'], label='High')
plt.plot(df['date'], df['low'], label='Low')

plt.title('Stock Prices Over A Decade')
plt.xlabel('Date')
plt.ylabel('Standardized Stock Price')
plt.legend()
plt.show()
```

```python
# Plot 1-2: Close, Open, High, Low
df = dataset[:10]

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label='Close', marker='o')
plt.plot(df['date'], df['open'], label='Open')
plt.plot(df['date'], df['high'], label='High')
plt.plot(df['date'], df['low'], label='Low')

plt.title('Stock Prices Over Last 10 days')
plt.xlabel('Date')
plt.ylabel('Standardized Stock Price')
plt.legend()
plt.show()
```

```python
# Plot 2: Close, GDP, EFFR, ACWI_Adjusted_Close
df = dataset

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['close'], label='S&P 500')
plt.plot(df['date'], df['gdp'], label='GDP')
plt.plot(df['date'], df['effr'], label='EFFR')
plt.plot(df['date'], df['acwi_adj_close'], label='ACWI')

plt.title('Trend of Stock Price and Economic Indicators Over A Decade')
plt.xlabel('Date')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.yticks([])  # Remove y-axis tick values
plt.legend()
plt.show()
```
